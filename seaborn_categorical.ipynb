```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#1
sns.stripplot(data=tips , x = 'day' , y = 'total_bill')
```

```python
#figure level funciton => catplot
sns.catplot(kind="strip",data=tips , x = 'day' , y = 'total_bill')
```

```python
#swarmplot
sns.swarmplot(data=tips , x = 'day' , y = 'total_bill')
```

```python
sns.catplot(kind="swarm",data=tips , x = 'day' , y = 'total_bill')
```

```python
sns.catplot(kind="swarm",data=tips , x = 'day' , y = 'total_bill' , hue = 'sex')
```

```python
#boxplot
sns.boxplot(data=tips , x = 'sex' , y = 'total_bill')
```

```python
sns.boxplot(data=tips , x = 'day' , y = 'total_bill')
```

```python
#using figure level function
sns.catplot(kind="box", data=tips, x='day', y='total_bill', hue='sex')
```

```python
#violinplot
sns.violinplot(data=tips , x = 'day' , y = 'total_bill')    
```
