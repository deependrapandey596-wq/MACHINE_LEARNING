## What is Seaborn?

- **Seaborn**: A Python library for data visualization, built on top of Matplotlib.
- **Purpose**: To make it easier to create beautiful and informative statistical graphics.

## Why Use Seaborn Over Matplotlib?

1. **Simpler Syntax**:
    - Seaborn requires less code to create complex visualizations.
    - Example: A line plot with Seaborn needs fewer lines of code than with Matplotlib.

2. **Better Default Styles**:
    - Seaborn comes with attractive default styles.
    - This means you get good-looking plots without much customization.

3. **Built-in Themes**:
    - Seaborn offers themes that can change the appearance of all plots easily.
    - Example: `sns.set_style("whitegrid")` applies a grid background to all plots.

4. **Integrated with Pandas**:
    - Seaborn works well with Pandas DataFrames.
    - This makes it easy to plot data directly from DataFrames.

5. **Advanced Statistical Plots**:
    - Seaborn provides functions for complex visualizations like heatmaps, violin plots, and pair plots.
    - These are either not available or harder to create in Matplotlib.

6. **Ease of Plot Customization**:
    - Seaborn simplifies the customization of plots.
    - Example: Changing colors, adding labels, and adjusting plot elements are more straightforward.

## 1. Relational plots.

Screenshot 2024-07-10 175256.png

```python
df = sns.load_dataset('penguins')   #load the penguins dataset
```

```python
df.sample(10)               #show 10 random rows of the dataset
```

### a. Scatterplot -> axes level function.      <!-- this helps     -->

```python
sns.relplot(data=df, x="flipper_length_mm", y="bill_length_mm")     #scatter plot of flipper length vs bill length
```

### hue parameter

```python
sns.relplot(data=df, x="flipper_length_mm", y="bill_length_mm",hue='species')   #hue is used to differentiate the points based on the species of penguin
```

### style parameter

```python
sns.relplot(data=df, x="flipper_length_mm", y="bill_length_mm",style='island')  #style is used to differentiate the points based on the island where the penguin was found
```

### size parameter

```python
sns.relplot(data=df, x="flipper_length_mm", y="bill_length_mm",size="body_mass_g")  #size is used to differentiate the points based on the body mass of the penguin. Larger points represent heavier penguins.
```

### All in One

```python
sns.scatterplot(data = df, x = 'flipper_length_mm', y = 'bill_length_mm', hue = 'species', style = 'island', size = 'body_mass_g')
```

## relplot -> figure level function
 - scatterplot

```python
sns.relplot(kind = 'line',data = df, x = 'flipper_length_mm', y = 'bill_length_mm', hue = 'species', style = 'island', size = 'body_mass_g')
```

## b. Lineplot -> axes level function

```python
sns.lineplot(data=df, x='flipper_length_mm', y='bill_length_mm')
```

### hue parameter

```python
sns.lineplot(data = df , x = 'flipper_length_mm' , y = 'bill_length_mm' ,hue='species')
```

### style parameter in line plot

```python
sns.lineplot(data = df , x = 'flipper_length_mm' , y = 'bill_length_mm' ,style='species')
```

```python
sns.lineplot(data = df , x = 'flipper_length_mm' , y = 'bill_length_mm' ,size='species')
```

Changing on values of parameter (hue, style and size) will effect the lineplot beacause they all are not as unique.

```python
sns.lineplot(data = df , x = 'flipper_length_mm' , y = 'bill_length_mm' ,hue='species' , style='species' , size='species')
```

## relplot -> figure level function
 - lineplot

```python
sns.relplot(kind='line' , data = df , x = 'flipper_length_mm' , y = 'bill_length_mm' ,hue='species' , style='species' , size='species')
```
