```python
sns.scatterplot(data = df, x = 'total_bill', y = 'tip',hue='day',style='sex')
#Parameters:
#hue = differentiate the color of the points based on the 'day' column
#style = differentiate the shape of the points based on the 'sex' column
#size = differentiate the size of the points based on a numerical column (not used in this example)
```

```python
#relplot is a higher level interface for drawing attractive and informative statistical graphics. It provides a convenient API for creating complex visualizations with multiple facets, allowing you to easily explore relationships between variables in your dataset.
sns.relplot(kind='scatter',data = df, x = 'total_bill', y = 'tip',hue='day',style='sex',size='size')
```
